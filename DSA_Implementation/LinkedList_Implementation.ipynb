# **Singly LinkedList Implementation**

```python
class Node:
    def __init__(self,item=None,next=None):
        self.item=item
        self.next=next

class SLL:
    def __init__(self,start=None):
        self.start=start
    def is_empty(self):
        return self.start==None
    def insert_at_start(self,item):
         n=Node(item,self.start)
         self.start=n
    def insert_at_last(self,item):
        n=Node(item)
        if self.is_empty():
            self.start=n
        else:
            curr=self.start
            while curr.next is not None:
                curr=curr.next
            curr.next=n
    def search(self,item):
        if self.start==None:
            return "Empty List"
        curr=self.start
        while curr is not None:
            if curr.item==item:
                return curr
            curr=curr.next
        return None 
    def insert_after(self,tmp,item):
        n=Node(item,tmp.next)
        if(tmp is not None):
            tmp.next=n
    def print_list(self):
        if self.start is None:
            print("Empty List")
        else:
            curr=self.start
            while curr:
                print(curr.item)
                curr=curr.next
    def delete_first(self):
        if self.start is None:
            print("Empty List")
        else:
            self.start=self.start.next
    def delete_last(self):
        if self.start is None:
            print("Empty List")
        elif self.start.next is None:
            self.start=None
        else:
            curr=self.start
            while curr.next.next is not None:
                curr=curr.next
            curr.next=None

    def delete_item(self,item):
        if self.start is None:
            print("Empty List")
        elif self.start.item==item:
            self.start=None
        elif self.start.next.item==item:
            self.start.next=None
        else:
            curr=self.start
            while curr.next.item!=item or curr!=None:
                curr=curr.next
            curr.next=curr.next.next

    def __iter__(self):
        return SLLIterable(self.start)


class SLLIterable:
    def __init__(self,start):
        self.current=start

    def __iter__(self):
        return self
    def __next__(self):
        if self.current is None:
            raise StopIteration
        else:
            data=self.current.item
            self.current=self.current.next
            return data
            
    
        
            
        
        
            

my_list=SLL()
my_list.is_empty()
my_list.insert_at_start(10)
my_list.insert_at_start(20)
my_list.insert_at_last(50)
my_list.print_list()
my_list.delete_first()
my_list.print_list()
print("........")
for i in my_list:
    print(i)
    
            
            

        
```

# **Doubly Linked List Implementation**

```python
class Node:
    def __init__(self,item,prev=None,next=None):
        self.item=item
        self.prev=prev
        self.next=next
class DLL:
    def __init__(self,start=None):
        self.start=start
    def is_empty(self):
        return self.start==None
    def printlist(self):
        if self.is_empty():
            return "Empty"
        else:
            curr=self.start
            while curr is not None:
                print(curr.item)
                curr=curr.next
    def insert_begin(self,item):
        n=Node(item)
        if self.start is not None:
            n.next=self.start
            self.start.prev=n
        self.start=n

    def insert_end(self,item):
        n=Node(item);
        if self.start is not None:
            curr=self.start
            while curr.next is not None:
                curr=curr.next
            curr.next=n
            n.prev=curr
        else:
            self.start=n
            
            


dll=DLL()
dll.insert_begin(10)
dll.insert_begin(20)
dll.insert_end(30)
dll.insert_begin(5)
dll.insert_end(50)
dll.printlist()


```
